```python
import matplotlib.pyplot as plt
```

```python
def logistic_func(x, r):
    res = r * x * (1 - x)
    return res
```

```python
f = lambda x : logistic_func(x, 4)
```

```python
print(f(0.5))
```

```python
# init 1 - x0 = 0.4
x01 = 0.2
x02 = 0.2001
y1 = [f(x01)]
y2 = [f(x02)]
N = 50
for i in range(N-1):
    y1.append(f(y1[-1]))
    y2.append(f(y2[-1]))
    
```

```python
print(y1)
print(y2)
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(N), y1, marker='.')
plt.plot(range(N), y2, marker='*')
```
