```python
import matplotlib.pyplot as plt
```

```python
def logistic_func(x, r):
    res = r * x * (1 - x)
    return res
```

```python
f1 = lambda x : logistic_func(x, 4)
```

```python
def get_series(f, x0, N):
    ys = [f(x0)]
    for _ in range(N-1):
        ys.append(f(ys[-1]))
    return ys
```

```python
# init 1 - x0 = 0.4

x01 = 0.2
x02 = 0.2001
N = 50
y1 = get_series(f1, x01, N)
y2 = get_series(f1, x02, N) 
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(N), y1, marker='.')
plt.plot(range(N), y2, marker='*')
```

```python
f2 = lambda x : logistic_func(x, 3.2)
```

```python
x03 = 0.2
x04 = 0.3
y3 = get_series(f2, x03, N)
y4 = get_series(f2, x04, N) 
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(N), y3, marker='.')
plt.plot(range(N), y4, marker='*')
```

**SDIC**: Sensitive Dependence on Initial Conditions. 

```python
N = 50
x01 = 0.3
e = 1e-6
x02 = x01 + e
r = 4.0
f = lambda x : logistic_func(x, r)
y1s = get_series(f, x01, N)
y2s = get_series(f, x02, N)
D_t = [abs(y1 - y2) for y1, y2 in zip(y1s, y2s)]
plt.figure(figsize=(12,6))
plt.plot(range(N), D_t)
```
